**Correlation Matrix**

**Model Training and Testing**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import pickle
```

```python
x= df.drop(columns= ['Species'])
y= df['Species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3)
```

```python
model= LogisticRegression()
model.fit(x_train, y_train)
```

```python
print("Logistic Regression Accuracy :" , model.score(x_test, y_test)*100)
```

```python
model.fit(x_train.values, y_train.values)
```

```python
print("Logistic Regression Accuracy :" , model.score(x_test, y_test)*100)
```

```python
modelk= KNeighborsClassifier()
modelk.fit(x_train, y_train)
```

```python
print("Logistic Regression Accuracy :" , modelk.score(x_test, y_test)*100)
```

```python
modelD= DecisionTreeClassifier()
modelD.fit(x_train, y_train)
```

```python
print("Logistic Regression Accuracy :" , modelD.score(x_test, y_test)*100)
```

**Saving the Model**

```python
import pickle
filename= 'model_KN.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
#save the updsted model

filename= 'model_KN.sav'
try:
    with open(filename, "wb") as file:
        pickle.dump(model, file)
    print("Model Saved Sucessfully")
except Exception as e:
    print(f"Error Saving the Model: {e}")
```

```python
load_model = pickle.load(open(filename,"rb"))
```
